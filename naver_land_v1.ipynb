```python
import requests
import json
import pandas as pd

headers= {
'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'
      } 
```

```python
#강남구 대치동
url = "https://m.land.naver.com/cluster/ajax/complexList?rletTpCd=APT&tradTpCd=A1&z=15&lat=37.4996377&lon=127.057867&btm=37.4902571&lft=127.0420741&top=37.5090171&rgt=127.0736598&showR0=&totCnt=66&cortarNo=1168010600"
res= requests.get(url, headers=headers)
```

```python
#크롤링한 데이터를 Json형태로 변환
data = json.loads(res.text)
items = data['result']
items[0]
```

```python
#변환된 Json데이터를 Pandas 도표화
df = pd.DataFrame.from_dict(items)
df = df.rename(columns={'hscpNm':'단지명','dealPrcMin': '매매최저가','dealPrcMax': '매매최고가','useAprvYmd': '승인날짜'})
df.columns
```

```python
#원하는 컴럼만 추출
Price_list = df[['단지명','승인날짜','매매최저가','매매최고가']]
Price_list
```
